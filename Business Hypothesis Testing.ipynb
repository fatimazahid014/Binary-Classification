### Loading data from dataset

```python
notes_data = pd.read_csv\
('https://raw.githubusercontent.com/fenago/datasets/main/BankNote_Authentication.csv')
#-------Type code here to display first five entries from daatset ------#


#------Solution------#
notes_data.head()
```

### Counting the number of records for each Skewness group

```python
# Write your code here:




# --- Solution ---#
filter_mask = notes_data['class'] == 0
notes = notes_data[filter_mask]\
           .groupby('skewness')['class'].agg(skewnessGrp='count')\
           .reset_index()
```

```python
 #Visualising the relationship using altair

alt.Chart(notes).mark_line().encode(x='skewness', y='skewnessGrp')
```

```python
# Getting another perspective
SkewnessTot = notes_data.groupby('skewness')['class']\
         .agg(SkewnessTot='count').reset_index()
SkewnessTot.head()
```

### Grouping the data by both skewness and class of notes and find the total counts under each category 

```python
# Getting all the details in one place
skewness_Final = notes_data.groupby(['skewness','class'])['class']\
          .agg(skewnessCat='count').reset_index()
skewness_Final.head()
```

```python
# Merging both the data frames

#Write your code here:



# --- Solution ---#
Comb = pd.merge(skewness_Final, SkewnessTot,left_on = ['skewness'], \
                  right_on = ['skewness'])
Comb['Category'] = (Comb.skewnessCat/Comb.SkewnessTot)*100
Comb.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
combYes = Comb[Comb['class'] == '0']
combNo = Comb[Comb['class'] == '1']
```

```python
xlabels = SkewnessTot['skewness'].nunique()
```

```python
Yes = combYes['skewnessCat'].unique()
No = combYes['skewnessCat'].unique()

Yes
No
```

```python
ind = np.arange(xlabels)
# Get the width of each bar
width = 0.35 
# Getting the plots
p1 = plt.bar(ind, Yes, width)
p2 = plt.bar(ind, No, width,bottom=Yes)
plt.ylabel('Authentication Measure(Proportion)')
plt.title('Authentication of Notes')
```

```python
plt.xticks(ind, SkewnessTot['Skew'].unique())
plt.yticks(np.arange(0, 100, 10)

plt.legend((p1[0], p2[0]), ('Yes', 'No'))
# To rotate the axis labels 
plt.xticks(rotation=90)
plt.show()
```
